```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import random
import decimal
import scipy.special
import math
from commpy.modulation import *
from commpy.impairments import *
from commpy.sequences import *
```

```python
# Randomize Data Bits
# Do this to eliminate long runs of 1's or 0's

def bit_randomizer(data):
    
    pn_seq_order = 12
    pn_seed = "110111011001"
    pn_mask = "010110101001"
    
    randomizer = pnsequence(pn_seq_order, pn_seed, pn_mask, len(data))
    
    randomized_data = data ^ randomizer
    
    return randomized_data
```

```python
# Modulate incoming data bits

def bpsk(data):
    mod_data = np.zeros(len(data), dtype=np.complex)
    constellation = [np.complex(-1, 0), np.complex(1, 0)]
    
    for i in range(len(data)):
        mod_data[i] = constellation[data[i]]
        
    return mod_data
            
def qpsk(data):
    qpsk_modem = QAMModem(4)
    return qpsk_modem.modulate(data) 

def sixteen_qam(data):
    qam16 = QAMModem(16)
    return qam16.modulate(data)
    
def sixfour_qam(data):
    qam64 = QAMModem(64)
    return qam64.modulate(data)

switch = {
    2: bpsk,
    4: qpsk,
    16: sixteen_qam,
    64: sixfour_qam
}

def modulate_data(data, psk_mod_order):
    func = switch.get(psk_mod_order, None)
    return func(data)    
    
```

```python
def generate_pilots(symbol_number, guard_left, dc_location, num_pilots, pilot_factor = 9):
    
    pilot_locations = np.zeros(num_pilots)
    
    for i in range(num_pilots):
        # vaugely wimax like
        loc = guard_left + (i * pilot_factor) + 3*(symbol_number % 3) + 1
        if (loc >= dc_location):
            pilot_locations[i] = loc + 1
        else:
            pilot_locations[i] = loc
            
    return pilot_locations
    
    
    
```

```python
# generate ofdm symbols
# pack data symbols into ofdm symbols and add pilot subcarriers, guard bands, and DC subcarrier
# returns a list of OFDM symbols
def generate_ofdm_symbol(fft_size, guard_left, guard_right, dc_location, pilots, data, pilot_modulation_order = 2):
    
    symbol = np.zeros(fft_size, dtype=np.complex)
    
    data_index = 0
    
    # seed the generator for random pilot generation
    np.random.seed(10)
    
    qpsk_constellation = [np.complex(1, 1), np.complex(1, -1), np.complex(-1, -1), np.complex(-1, 1)]
    
    for i in range(guard_left, fft_size - guard_right):
        if (i != dc_location):
            if (i in pilots):
                # No pilot modulation
                if (pilot_modulation_order == 1):
                    symbol[i] = np.complex(1, 0) # All pilots are set to 1,0
                
                # BPSK pilot modulation order
                elif (pilot_modulation_order == 2):
                    symbol[i] = np.complex(np.random.randint(0,2) * 2 - 1, 0)
                
                # QPSK pilot modulation order
                elif (pilot_modulation_order == 4):
                    temp = np.random.randint(0,4)
                    symbol[i] = qpsk_constellation[temp]
                           
            else:
                symbol[i] = data[data_index]
                data_index += 1
        
    
    return symbol
    
```

```python


def ofdm_transmitter(fft_size, sample_rate, ofdm_factor, pilot_factor, pn_seq_order, pn_seed, pn_mask, modulation_order):

    pn_seq = pnsequence(pn_seq_order, pn_seed, pn_mask, 2**(pn_seq_order))
    guard_left = (fft_size / ofdm_factor)/2
    guard_right = guard_left - 1

    dc_location = (fft_size / 2) - 1

    num_usable_subcarriers = fft_size - guard_left - guard_right - 1

    num_pilots = num_usable_subcarriers/pilot_factor

    num_data_subcarriers = num_usable_subcarriers - num_pilots
    
    num_data_symbols = len(pn_seq)/ math.log(modulation_order, 2)

    pad_len = math.log(modulation_order, 2) * (num_data_subcarriers - (num_data_symbols % num_data_subcarriers))
    
    data = np.pad(pn_seq, (0,int(pad_len)), 'constant')
    
    num_symbols = int(len(data)/ math.log(modulation_order, 2)) / num_data_subcarriers

    print ("OFDM Parameters:\nFFT Size: %d\nSample Rate: %d\nOFDM Factor: %d\nPilot Factor: %d" % (fft_size, sample_rate, ofdm_factor, pilot_factor))
    print ("Modulation Order: %d" % modulation_order)
    print ("Guard Left: %d\nGuard Right: %d\nDC Location: %d" %(guard_left, guard_right, dc_location))
    print ("Num Pilots: %d\nNum Data Subcarriers: %d\nData Len (bits): %d\nPad Len (bits): %d" % (num_pilots, num_data_subcarriers, len(pn_seq), pad_len))
    print ("Num Symbols: %d" % num_symbols)
    
    randomized_data = bit_randomizer(data)
    
    modulated_data = modulate_data(randomized_data, modulation_order)
    
    ofdm_symbols = []
    
    for i in range(num_symbols):
        pilot_locs = generate_pilots(i, guard_left, dc_location, num_usable_subcarriers, pilot_factor)
        raw_symbol = generate_ofdm_symbol(fft_size, guard_left, guard_right, dc_location, pilot_locs, modulated_data)
        ofdm_symbols.append((raw_symbol))
        
    return ofdm_symbols

```

```python
# Set the parameters for the OFDM transmitter

fft_size = 1024
sample_rate = 10240000
ofdm_factor = 8
pilot_factor = 8
modulation_order = 64

pn_seq_order = 13
pn_seed = "1010111101101"
pn_mask = "1111010111011"

symbols = ofdm_transmitter(fft_size, sample_rate, ofdm_factor, pilot_factor, pn_seq_order, pn_seed, pn_mask, modulation_order)

plt.figure()
plt.ylim(-math.log(modulation_order, 2), math.log(modulation_order, 2))
plt.xlim(-math.log(modulation_order, 2), math.log(modulation_order, 2))
plt.title("Constellation Plot")

for i in symbols:
    plt.plot(np.real(i), np.imag(i), 'ro')
    
    
```
