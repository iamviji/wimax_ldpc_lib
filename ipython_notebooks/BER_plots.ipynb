```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import random
import decimal
import scipy.special
import math
```

<center>**OFDM Subcarrier Spacing:**</center>
<br/>
$$\textrm{Subcarrier Spacing (Hz)} = \frac{F_s}{N_{FFT}}$$

<center>**OFDM Channel Capacity:**</center>
<br/>
$$C_{OFDM} = \frac{1}{\ln{2}} \exp \Big(\frac{N_0}{2 P_0 T_s}\Big) E_1 \Big( \frac{N_0}{2 P_0 T_s}\Big)$$

<center>**Equation for SNR:**</center>
<br/>
$$ SNR = \frac{Signal Power}{Noise Power} = \frac{E_s \times F_s}{N_0 \times B}$$
<br/>
$$ SNR = \Big(\frac{F_s}{B}\Big) \Big(\frac{E_s}{N_0}\Big) = N_b \Big(\frac{F_s}{B}\Big) \Big(\frac{E_b}{N_0}\Big)\\ \\ \textrm{where:}\\
F_s = \textrm{symbol rate (1/sec)} \\
B = \textrm{bandwidth (Hz = 1/sec)} \geq F_s \\
N_0 = \textrm{noise power spectral density} \\
E_s = \textrm{Energy per symbol (J)}\\
E_b = \textrm{Energy per bit (J)} \\
N_b = \textrm{bits per symbol}$$
<br/>


<center>**Thermal Noise:**</center>
<br/>
$$ N_0 = kT = (1.383 \times 10^{-23} J / K)\times T $$

<center> **Eb/N0 and SNR for MQAM Signals:**</center>
<br/>
$$ SNR = \frac{E_s / T_s}{N_0 B} = \Big( \frac{\log_2{(M)} E_b}{N_0} \Big) \Big(\frac{F_s}{B} \Big) \\
\textrm{where:}\\
M = \textrm{# of bits per symbol}\\$$
$$\textrm{therefore, if } F_s \cong B\\
\frac{E_b}{N_0} = \Big( \frac{1}{\log_2{(M)}}\Big) SNR$$

<center> ** Probability of Symbol Error in AWGN: **</center>
<br\>
$$ \textrm{BPSK:    } P_s = Q\Bigg(\sqrt{2 \frac{E_b}{N_0}}\Bigg) $$
$$ \textrm{QPSK:   } P_s \approx 2Q\Bigg(\sqrt{2 \frac{E_b}{N_0}}\Bigg) $$
$$ \textrm{M-QAM:   } P_s \approx 4Q\Bigg(\sqrt{\bigg(\frac{3\log_2{M}}{M-1}\bigg) \bigg(\frac{E_b}{N_0}\bigg)}\Bigg) $$
<br\>
$$ \textrm{where:} \\$$
$$ Q(x) = \frac{1}{\sqrt{2\pi}} \int_{x}^{+\infty} e^{-t^2 dt}$$
$$ = \frac{1}{2}erfc\Bigg(\frac{x}{\sqrt{2}}\Bigg) $$

<center> ** Probability of Bit Error with Gray Coding: **</center>
<br\>
$$ \textrm{M-QAM:    } P_b \approx \frac{1}{\log_2{(M)}} P_s \\$$


<center>**Channel Capacity:**</center>
$$ C = B \log_2{1 + SNR} \textrm{   bits/sec} $$

```python
def Q(x):
    return 0.5 * scipy.special.erfc(x / math.sqrt(2))

EbNodB = np.arange(-10,25)
bpsk = np.zeros(len(EbNodB))
qpsk = np.zeros(len(EbNodB))
sixteenqam = np.zeros(len(EbNodB))
sixfourqam = np.zeros(len(EbNodB))

for i in range(len(EbNodB)):
    EbNo = 10.0 ** (EbNodB[i]/10.0)
    bpsk[i] = Q(math.sqrt(2*EbNo))
    qpsk[i] = 2*Q(math.sqrt(2*EbNo))
    sixteenqam[i] = 4 * Q(math.sqrt(((3 * math.log(16,2))/(16 - 1))*EbNo))
    sixfourqam[i] = 4 * Q(math.sqrt(((3 * math.log(64,2))/(64 - 1))*EbNo))
    
print type(bpsk)
```

```python
fig, axes = plt.subplots(nrows =1, ncols=2, figsize=(15,6))

plt.suptitle("SER and BER Curves for select modulation schemes in AWGN\n", fontsize=16)
plt.subplots_adjust(top=0.85)
axes[0].plot(EbNodB, bpsk, label='BPSK')
axes[0].plot(EbNodB, qpsk, label='QPSK')
axes[0].plot(EbNodB, sixteenqam, label='16-QAM')
axes[0].plot(EbNodB, sixfourqam, label='64-QAM')

axes[0].axis([-5, 25, 1e-8, .1])
axes[0].set_yscale('log')
axes[0].grid(True)
axes[0].set_xlabel('EbNo(dB)')
axes[0].set_ylabel('SER')
axes[0].set_title('Symbol Error Rate Curves')
axes[0].legend(loc=3)

axes[1].plot(EbNodB, ((1.0)/(math.log(2, 2))) * bpsk, label='BPSK')
axes[1].plot(EbNodB, ((1.0)/(math.log(4, 2))) * qpsk, label='QPSK')
axes[1].plot(EbNodB, ((1.0)/(math.log(16, 2))) * sixteenqam, label='16-QAM')
axes[1].plot(EbNodB, ((1.0)/(math.log(64, 2))) * sixfourqam, label='64-QAM')

axes[1].axis([-5, 25, 1e-8, .1])
axes[1].set_yscale('log')
axes[1].grid(True)
axes[1].set_xlabel('EbNo(dB)')
axes[1].set_ylabel('BER')
axes[1].set_title('Bit Error Rate Curves')
axes[1].legend(loc=3)
```
